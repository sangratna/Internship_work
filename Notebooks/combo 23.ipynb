```python
df_new=df[['Date', 'Time', 'Minutes', 'PA Home PPM', 'PA Home temprature',
       'PA Home pressure(mbar)', 'PA Home RH', 'PA Home DPT', 'PA Home AH','Outdoor PPM', 'Outdoor Temp',
       'Outdoor Pressure(mbar)', 'Outdoor RH', ' Outdoor DPT', 'Outdoor AH']]
```

```python
# Convert the columns to numeric data type
import warnings
warnings.filterwarnings("ignore")
df_new['Outdoor PPM'] = pd.to_numeric(df_new['Outdoor PPM'], errors='coerce')
df_new['PA Home PPM'] = pd.to_numeric(df_new['PA Home PPM'], errors='coerce')

df_new['Outdoor Temp'] = pd.to_numeric(df_new['Outdoor Temp'], errors='coerce')
df_new['PA Home temprature'] = pd.to_numeric(df_new['PA Home temprature'], errors='coerce')

df_new['Outdoor Pressure(mbar)'] = pd.to_numeric(df_new['Outdoor Pressure(mbar)'], errors='coerce')
df_new['PA Home pressure(mbar)'] = pd.to_numeric(df_new['PA Home pressure(mbar)'], errors='coerce')

df_new['Outdoor RH'] = pd.to_numeric(df_new['Outdoor RH'], errors='coerce')
df_new['PA Home RH'] = pd.to_numeric(df_new['PA Home RH'], errors='coerce')

df_new[' Outdoor DPT'] = pd.to_numeric(df_new[' Outdoor DPT'], errors='coerce')
df_new['PA Home DPT'] = pd.to_numeric(df_new['PA Home DPT'], errors='coerce')

df_new['PA Home AH'] = pd.to_numeric(df_new['PA Home AH'], errors='coerce')
df_new['Outdoor AH'] = pd.to_numeric(df_new['Outdoor AH'], errors='coerce')
```

```python
df_new['PPM_diff']=df_new['PA Home PPM']-df_new['Outdoor PPM']
df_new['Temp_diff']=df_new['Outdoor Temp']-df_new['PA Home temprature']
# df_new['Press_diff']=df_new['Outdoor Pressure(mbar)']-df_new['Exp chamber pressure']
df_new['RH_diff']=df_new['Outdoor RH']-df_new['PA Home RH']
df_new['DPT_diff']=df_new[' Outdoor DPT']-df_new['PA Home DPT']
df_new['AH_diff']=df_new['Outdoor AH']-df_new['PA Home AH']
```

```python
df_new.keys()
```

```python
df_diff=df_new[['Outdoor PPM', 'Outdoor Temp', 'Outdoor Pressure(mbar)', 'Outdoor RH',
       ' Outdoor DPT', 'Outdoor AH', 'PPM_diff', 'Temp_diff', 'RH_diff',
       'DPT_diff', 'AH_diff']]
```

```python
df_diff.isnull().sum()
```

```python
A=df_diff['Outdoor PPM']
B=df_diff[['PPM_diff']]
import warnings
warnings.filterwarnings('ignore')
from sklearn.cluster import KMeans
wcss=[]
for k in range(1,11):
    kmeans=KMeans(n_clusters=k,init='k-means++')
    kmeans.fit(B)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),wcss)
plt.xticks(range(1,11))
plt.xlabel("Number of Clustrers in PPM_diff")
plt.ylabel("WCSS")
plt.show()
```

```python
sns.scatterplot(x=df_diff['Outdoor PPM'],y=df_diff['PPM_diff'])
```

```python
kmeans=KMeans(n_clusters=3,init="k-means++")
y_labels=kmeans.fit_predict(B)
plt.scatter(A,B,c=y_labels,marker='o',edgecolors='black')
plt.xlabel('Outdoor PPM')
plt.ylabel('PPM_diff')
```

```python
import pandas as pd

# Creating a dictionary to store indices for each cluster
cluster_indices = {}
for i, label in enumerate(y_labels):
    if label in cluster_indices:
        cluster_indices[label].append(i)
    else:
        cluster_indices[label] = [i]

# Creating a list to store index and cluster number
data = []
for cluster_label, indices in cluster_indices.items():
    for index in indices:
        data.append({'Index': index, 'Cluster': cluster_label})

# Creating a DataFrame from the list of dictionaries
df = pd.DataFrame(data)

# Exporting the DataFrame to an Excel file with the specified path
file_path = r'C:\Users\good day\Desktop\ppm_cluster_data.xlsx'
df.to_excel(file_path, index=False)
```
