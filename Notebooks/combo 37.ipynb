```python
df_new=df[['Minutes','Open office far from AC PPM','Open office far from AC Temprature',
'Open office far from AC Pressure(mbar)', 'Open office far from AC RH',
           'Open office far from AC DPT', 'Open office far from AC AH','Outdoor PPM', 'Outdoor Temp',
       'Outdoor Pressure(mbar)', 'Outdoor RH', ' Outdoor DPT', 'Outdoor AH']]
```

```python
df_new.isnull().sum()
```

```python
df_new = df_new.drop(df[df.eq('###').any(axis=1)].index)
```

```python
df_new.shape
```

```python
df_new['Outdoor PPM'] = pd.to_numeric(df_new['Outdoor PPM'], errors='coerce')
df_new['Open office far from AC PPM'] = pd.to_numeric(df_new['Open office far from AC PPM'], errors='coerce')

df_new['Outdoor Temp'] = pd.to_numeric(df_new['Outdoor Temp'], errors='coerce')
df_new['Open office far from AC Temprature'] = pd.to_numeric(df_new['Open office far from AC Temprature'], errors='coerce')

df_new['Outdoor Pressure(mbar)'] = pd.to_numeric(df_new['Outdoor Pressure(mbar)'], errors='coerce')
df_new['Open office far from AC Pressure(mbar)'] = pd.to_numeric(df_new['Open office far from AC Pressure(mbar)'], errors='coerce')

df_new['Outdoor RH'] = pd.to_numeric(df_new['Outdoor RH'], errors='coerce')
df_new['Open office far from AC RH'] = pd.to_numeric(df_new['Open office far from AC RH'], errors='coerce')

df_new[' Outdoor DPT'] = pd.to_numeric(df_new[' Outdoor DPT'], errors='coerce')
df_new['Open office far from AC DPT'] = pd.to_numeric(df_new['Open office far from AC DPT'], errors='coerce')

df_new['Outdoor AH'] = pd.to_numeric(df_new['Outdoor AH'], errors='coerce')
df_new['Open office far from AC AH'] = pd.to_numeric(df_new['Open office far from AC AH'], errors='coerce')
```

```python
df_new['PPM_diff']=df_new['Open office far from AC PPM']-df_new['Outdoor PPM']
df_new['Temp_diff']=df_new['Outdoor Temp']-df_new['Open office far from AC Temprature']
df_new['Press_diff']=df_new['Outdoor Pressure(mbar)']-df_new['Open office far from AC Pressure(mbar)']
df_new['RH_diff']=df_new['Outdoor RH']-df_new['Open office far from AC RH']
df_new['DPT_diff']=df_new[' Outdoor DPT']-df_new['Open office far from AC DPT']
df_new['AH_diff']=df_new['Outdoor AH']-df_new['Open office far from AC AH']
```

```python
df_new.keys()
```

```python
A=df_new['Outdoor PPM']
B=df_new[['PPM_diff']]
import warnings
warnings.filterwarnings('ignore')
from sklearn.cluster import KMeans
wcss=[]
for k in range(1,11):
    kmeans=KMeans(n_clusters=k,init='k-means++')
    kmeans.fit(B)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),wcss)
plt.xticks(range(1,11))
plt.xlabel("Number of Clustrers in PPM_diff")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans=KMeans(n_clusters=3,init="k-means++")
y_labels=kmeans.fit_predict(B)
plt.scatter(A,B,c=y_labels,marker='o',edgecolors='black')
plt.xlabel('Outdoor PPM')
plt.ylabel('PPM_diff')
```

```python
import pandas as pd

# Creating a dictionary to store indices for each cluster
cluster_indices = {}
for i, label in enumerate(y_labels):
    if label in cluster_indices:
        cluster_indices[label].append(i)
    else:
        cluster_indices[label] = [i]

# Creating a list to store index and cluster number
data = []
for cluster_label, indices in cluster_indices.items():
    for index in indices:
        data.append({'Index': index, 'Cluster': cluster_label})

# Creating a DataFrame from the list of dictionaries
df = pd.DataFrame(data)

# Exporting the DataFrame to an Excel file with the specified path
file_path = r'C:\Users\good day\Desktop\PPM_cluster_data.xlsx'
df.to_excel(file_path, index=False)
```
